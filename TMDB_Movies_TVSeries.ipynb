We have cast data as shown below but the desired data we need is in lists

To create a list we will define a function, that will create a list which contains name of actors from the available json format.

```python
def fetch_actors(obj):
    L = []
    for i in obj:
        if i['known_for_department'] == 'Acting':
            L.append(i['name'])
    return L
```

```python
# using apply function to apply our defined function 
df['cast'] = df['cast'].apply(fetch_actors)
```

To create a list we will define a function, that will create a list which contains name of director from the available json format.

```python
def fetch_director(obj):
    L = []
    for i in obj:
        if i['job'] == 'Director':
            L.append(i['name'])
    return L
```

```python
# using apply function to apply our defined function 
df['crew'] = df['crew'].apply(fetch_director)
```
